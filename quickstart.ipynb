### Embeddings Example ###
Note that the length of the embedding vector is fixed and does not depend on the length of the input string.  Using the model `text-embedding-ada-002`, the embedding vectore size is 1536 as indicated in the code two cells below.

```python
import os
import openai
openai.api_key = os.getenv("OPENAI_API_KEY")
response = openai.Embedding.create(
  model="text-embedding-ada-002",
  input="There are four sides to a square.",
)
print(response)
```

```python
embedding_length=len(response.data[0].embedding)
print(embedding_length)
```
